```python
data = pd.read_csv('C:/Users/Amiya/Desktop/eZmFnBJAmP.csv')
data = data.sort_values(by='Name of the Political Party')
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,6))
```

```python
party_totals.plot(kind='bar')
```

```python
party_totals = data.groupby('Name of the Political Party')['Denomination'].sum()
party_totals = party_totals.sort_values()
# Threshold for grouping small values
threshold = 0.05  # For example, grouping values less than 5%

# Filter out small values and sum them
small_values_sum = party_totals[party_totals < party_totals.sum() * threshold].sum()

# Filter out large values
party_totals_filtered = party_totals[party_totals >= party_totals.sum() * threshold]

# Add the small values sum to the "Others" category
party_totals_filtered['Others'] = small_values_sum

# Plotting
plt.figure(figsize=(8, 8))  # Set the figure size
plt.pie(party_totals_filtered.values, labels=party_totals_filtered.index, autopct='%1.1f%%', startangle=140)
plt.title('Total Amount Encashed by Political Party')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.tight_layout()  # Adjust layout
plt.show()
```
